```python
#1. Download the A2/data folder
class DownloadProgressBar(tqdm):
    def update_to(self, b=1, bsize=1, tsize=None):
        if tsize is not None:
            self.total = tsize
        self.update(b * bsize - self.n)

def download_url(download_path: Path, url: str):
    with DownloadProgressBar(unit='B', unit_scale=True,
                             miniters=1, desc=url.split('/')[-1]) as t:
        urllib.request.urlretrieve(url, filename=download_path, reporthook=t.update_to)
```

```python
def download_dataset(download_path: Path, url: str):
    print("Downloading dataset...")
    download_url(url=url, download_path=download_path)
    print("Download complete!")


def extract_dataset(download_path: Path, extract_path: Path):
    print("Extracting dataset... (it may take a while...)")
    # Check if the file is a ZIP file
    if zipfile.is_zipfile(download_path):
        with zipfile.ZipFile(download_path, 'r') as zip_ref:
            zip_ref.extractall(extract_path)
        print("Extraction completed!")
    else:
        print("Error: The downloaded file is not a ZIP file.")
```

## Prompt Template

Use the following prompt template to process input texts.

```python
# Use the dataframe we have created before to get the texts which we will clasify later
texts = a2_df['text'].tolist()
```

```python
prompt = [
    {
        'role': 'system',
        'content': 'You are an annotator for sexism detection.'
    },
    {
        'role': 'user',
        'content': """Your task is to classify input text as containing sexism or not. Respond only YES or NO.

        TEXT:
        {text}

        ANSWER:
        """
    }
]
```

### Instructions

In order to get Task 2 points, we require you to:

* Write a ``prepare_prompts`` function as the one reported below.

```python
def prepare_prompts(texts, prompt_template, tokenizer):
    """
    This function formats input text samples into instruction prompts.

    Inputs:
      texts: list of input texts to classify via prompting
      prompt_template: the prompt template provided in this assignment
      tokenizer: the transformers Tokenizer object instance associated with the chosen model card

    Outputs:
      List of tokenized instruction prompts ready for model inference.
    """
    tokenized_prompts = []

    for text in texts:
        formatted_prompt = [
            {
                'role': prompt_template[0]['role'],
                'content': prompt_template[0]['content']
            },
            {
                'role': prompt_template[1]['role'],
                'content': prompt_template[1]['content'].format(text=text)
            }
        ]

        # Convert the formatted prompt into a single string for tokenization
        prompt_str = f"{formatted_prompt[0]['content']}\n\n{formatted_prompt[1]['content']}"

        # Tokenize the formatted prompt
        tokenized_prompt = tokenizer(
            prompt_str,
            return_tensors="pt",
            padding=True,
            truncation=True
        )

        tokenized_prompts.append(tokenized_prompt)

    return tokenized_prompts
```

```python
prepared_prompts = prepare_prompts(texts, prompt, tokenizer)

for idx, prompt in enumerate(prepared_prompts):
    print(f"Prompt {idx + 1}:")
    print(prompt)
```

```python
# Ensure the inputs are on the same device as the model
device = model.device

input_ids = prepared_prompts[0]['input_ids'].to(device)
attention_mask = prepared_prompts[0]['attention_mask'].to(device)

# Generate output
outputs = model.generate(
    input_ids=input_ids,
    attention_mask=attention_mask,
    max_new_tokens=50
)

# Decode the output
response = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(response)

```

```python
# Ensure the inputs are processed in a loop
device = model.device

# List to store the decoded responses
responses = []

# Iterate over all prompts
for idx, prompt in enumerate(prepared_prompts):
    input_ids = prompt['input_ids'].to(device)
    attention_mask = prompt['attention_mask'].to(device)

    # Generate the output
    outputs = model.generate(
        input_ids=input_ids,
        attention_mask=attention_mask,
        max_new_tokens=50
    )

    # Decode the output
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    responses.append(response)

    # Print each response
    #print(f"Prompt {idx + 1} Response: {response}")

# Optionally, save all responses to a file
with open("responses.txt", "w") as file:
    for idx, response in enumerate(responses):
        file.write(f"Prompt {idx + 1} Response: {response}\n")
```
